```python
import math

#(Base algorithm from the book page 369, Extended-Bottom-Up-Cut-Rod) Returns the revenue array r and the optimal cut array s. C is the cost per cut.
def extendedBottomUpCutRod(p,n,c):
  #Initializes arrays r and s, the revenue and solution arrays.
  r = [0] * (n+1)
  s = [0] * (n+1)
  #Solves subproblems of size j.
  for j in range(1, n + 1):
    q = (-math.inf)
    for i in range(1, j + 1):
      #Sets solution to the subproblem and the optimal cut
      if q < (p[i] + r[j-i]):
        q = p[i] + r[j-i] - c
        s[j] = i
      #Saves solution to the subproblem of size j.
      r[j] = q
  return r,s

#(Base algorithm from the book page 369, Print-Cut-Rod-Solution) Returns the optimal cut solution.
def printCutRodSolution(p,n,c):
  #Sets array a to array r and b to s from the extendedBottomUpCutRod
  a,b = extendedBottomUpCutRod(p, n, c)
  #Returns the optimal solution.
  while n > 0:
    print(b[n])
    n = n - b[n]
  
#Test values
valueList = [0, 8, 14, 30, 33, 42, 50]
arrayLength = len(valueList)
cutCost = 5

#Function calls
print(extendedBottomUpCutRod(valueList, arrayLength - 1, cutCost), printCutRodSolution(valueList, arrayLength - 1, cutCost))


```
